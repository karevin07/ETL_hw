#  爬蟲
## 找第1頁

```python
def get_total_page_cnt():
    URL = "https://www.ptt.cc/bbs/Soft_Job/index.html"
    resp = requests.get(URL)

# 這個符號
# ‹-> &lsaquo;
    total_page_cnt = int(re.findall('href="/bbs/Soft_Job/index(\d+).html">&lsaquo; 上頁', resp.text)[0]) + 1

    return total_page_cnt 
```

```python
#找出第1頁的INDEX
get_total_page_cnt()
```

```python
def get_list_page(url):
    HOST = "https://www.ptt.cc"
    resp = requests.get(url)
    links = re.findall('<a href="(/bbs/Soft_Job/M.+\.html)">.徵才..+</a>', resp.text)
    detail_page_links = [HOST + link for link in links]
    return detail_page_links
```

```python
def get_list(url):
    filename = "_".join(url.split('/')[-1].split('.')[:-1]) + '.html'
    filelist.append(filename)
```

```python
def dump_page(url):
    filename = "_".join(url.split('/')[-1].split('.')[:-1]) + '.html'
    
    custom_headers = {
        "cookie": "over18=1;"
    }

    resp = requests.get(url, headers=custom_headers)
    with open(DATA_PATH + '/' + filename, 'w') as f:
        f.write(resp.text)
    return filename
```

```python
def get_link(num):
    print('共爬'+str(num)+'頁')
    URL_TEMPLATE = "https://www.ptt.cc/bbs/Soft_Job/index{}.html"
    total_page_cnt = get_total_page_cnt()
    count = 0
    start = time.time()
    for pg in range(total_page_cnt, total_page_cnt -num, -1):
        url = URL_TEMPLATE.format(pg)
        for link in get_list_page(url):
            dump_page(link)
            count+=1
            get_list(link)
        end = time.time()
    elapsed = end - start
    print('共'+str(count)+'筆職缺')
    print("花費", elapsed, "秒")
```

# 內文清理

```python
for l in filelist:   
    with open(DATA_PATH + '/' + l) as f:
        html = f.read()
        soup = BeautifulSoup(html, 'lxml')
        content = soup.select('div#main-content ')
        for a in range(len(content)):
            words.append(content[a].text.lower())
```

```python
allword = []
for b in range(len(words)):
    allword.append(words[b])
```

```python
allwords = []
thewords = jieba.cut(allword, cut_all=False)
for word in thewords:
    allwords.append(word)
allwords
```

```python
#建技能表
the_skill = []

with open('data/skill.txt') as sk:
    skilllist =  sk.read()
skill = skilllist.split('\n')
skill.remove('')
for ski in range(len(skill)):
    s = skill[ski].lower()
    the_skill.append(s)
```

```python
# 計次歸零
counter = {}
for b in the_skill:
    counter[b]=0
```

```python
counter
```

```python
# 計次
def get_skill(list):    
    for a in range(len(list)):
        word = list[a]
        if word in the_skill:                            
            counter[word]+=1
    return counter
```
