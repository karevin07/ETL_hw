#  爬蟲
## 找第1頁

```python
def get_total_page_cnt():
    URL = "https://www.ptt.cc/bbs/Soft_Job/index.html"
    resp = requests.get(URL)

# 這個符號
# ‹-> &lsaquo;
    total_page_cnt = int(re.findall('href="/bbs/Soft_Job/index(\d+).html">&lsaquo; 上頁', resp.text)[0]) + 1

    return total_page_cnt 
```

```python
#找出第1頁的INDEX
get_total_page_cnt()
```

```python
def get_list_page(url):
    HOST = "https://www.ptt.cc"
    resp = requests.get(url)
    links = re.findall('<a href="(/bbs/Soft_Job/M.+\.html)">.+</a>', resp.text)
    detail_page_links = [HOST + link for link in links]
    return detail_page_links
```

```python
def get_list(url):
    filename = "_".join(url.split('/')[-1].split('.')[:-1]) + '.html'
    filelist.append(filename)
```

```python
def get_link(num):
    print('共爬'+str(num)+'頁')
    URL_TEMPLATE = "https://www.ptt.cc/bbs/Soft_Job/index{}.html"

    total_page_cnt = get_total_page_cnt()

    for pg in range(total_page_cnt, total_page_cnt -num, -1):
        url = URL_TEMPLATE.format(pg)
        for link in get_list_page(url):
            dump_page(link)
            get_list(link)
```

```python
def dump_page(url):
    filename = "_".join(url.split('/')[-1].split('.')[:-1]) + '.html'
    
    custom_headers = {
        "cookie": "over18=1;"
    }

    resp = requests.get(url, headers=custom_headers)
    with open(DATA_PATH + '/' + filename, 'w') as f:
        f.write(resp.text)
    return filename
```
