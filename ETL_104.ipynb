```python
#總共幾頁列表
def get_totalpage(jobname):
    url = job_url.format(jobname)
    total = int(re.findall('"TOTALPAGE":"(\d+)"',requests.get(url).text)[0])
    return total
```

# 抓出單頁


```python
def get_list_page(jobname):
    for page in range(1,100 ):
        url = job_url.format(jobname)
        resp = requests.get(url + '&page=' + str(5))
        job_data = resp.json()['data']
        jobs.extend(job_data)                               
        time.sleep(1.5)                                     
```

```python
get_list_page('軟體工程師')
```

```python
def dump_page(url):  
    filename = re.split('&|\?|=',url)[2]
    resp = requests.get(url)
    with open(DATA_PATH + '/' + filename, 'w') as f:
        f.write(resp.text)
    return filename
```

```python
def get_list(url):
    filename = re.split('&|\?|=',url)[2]
    filelist.append(filename)
```

```python
def do_page(num1 ,a):   
    print('職缺 '+a)
    print('共抓'+str(num1)+'頁')
    count = 0
    filelist = []
    start = time.time()
    for page in range(num1) :
        the_url = job_url.format(a,page+1)
        for link in get_list_page(the_url):
            dump_page(link)
            count+=1
            get_list(link)
    end = time.time()
    elapsed = end - start
    print('共'+str(count)+'筆職缺')
    print("花費", elapsed, "秒")
```

```python
do_page(20,'程式設計師')
```

# 輸出Json 、 CSV

```python
def get_json(dict,file):
    with open(file, 'w') as f:
        json.dump(dict,f)
```

```python
def get_csv(dict,file):
    with open(file,'w') as f:
        w=csv.writer(f)
        # write all keys on one row and all values on the next
        w.writerow(dict.keys())
        w.writerow(dict.values())
```

```python
def get_tupletocsv(list,file):

    with open(file,'w') as out:
        csv_out=csv.writer(out)
        csv_out.writerow(['技能','數量'])
        for row in list:
            csv_out.writerow(row)
```
